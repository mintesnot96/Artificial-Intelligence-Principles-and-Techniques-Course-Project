#### 1.2 Write a class that takes the converted state space graph, initial state, goal state and a search strategy and return the corresponding solution/path according to the given strategy. Please consider only breadth-first search and depth-first search strategies for this question.

```python
class Search:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal
        
    def bfs(self):
        queue = [(self.start, [self.start])]
#         print(queue)
        visited = set()
        while queue:
            node, path = queue.pop(0)
            
            if node == self.goal:
                print('\n\n\n Result')  
                return path
            if node not in visited:
                visited.add(node)
                for neighbor in self.graph[node]:
                    queue.append((neighbor, path + [neighbor]))
#                     print("33333333333333")
#                     print(neighbor,'=====+======',path)
                    
        return None
    
    def dfs(self):
        stack = [(self.start, [self.start])]
        visited = set()
        while stack:
            node, path = stack.pop()
            if node == self.goal:
                print('\n\n\n Result')  
                return path
            
            if node not in visited:
                visited.add(node)
                for neighbor in reversed(self.graph[node]):
                    stack.append((neighbor, path + [neighbor]))
        return None

def maindf(a,b,c):
    search =Search(graph, b, c)
    if a=='bfs':      
        print('\n BFS')
        print(search.bfs()) 
    else:
        print('\n')
        print('\n DFS ')
        print(search.dfs())
    
a=input('Strategy bfs or dfs: ')
b=input('Initial State: ')
c=input('Goal State: ')
print(maindf(a,b,c))
```

#### 2.2.  Assuming â€œAddis Ababaâ€ as an initial state, write a program that use uniform cost search to generate a path to â€œLalibelaâ€.

```python
from queue import PriorityQueue

import heapq
def uniform_cost_search(graph, initial_state, goal_state):

    frontier = PriorityQueue()
    frontier.put((0, initial_state))
#     print(frontier)
    visited = set()

    cost_so_far = {}
    cost_so_far[initial_state] = 0
    path = {}
    path[initial_state] = []

    while not frontier.empty():
        current_cost, current_state = frontier.get()

        if current_state == goal_state:
            return path[current_state] + [current_state]

        visited.add(current_state)

        for neighbor, cost in graph[current_state].items():
            new_cost = cost_so_far[current_state] + cost
            if neighbor not in visited or new_cost < cost_so_far[neighbor]:
                cost_so_far[neighbor] = new_cost
                priority = new_cost
                frontier.put((priority, neighbor))
                path[neighbor] = path[current_state] + [current_state]
    return None

uniform_cost_search(graph_ETUCS, 'AddisAbaba', 'Lalibela')
```

#### 2.3. Given â€œAddis Ababaâ€ as an initial state and â€œAxumâ€, â€œGondarâ€, â€œLalibelaâ€, Babile, â€œJimmaâ€, â€œBaleâ€, â€œSof Oumerâ€, and â€œArba Minchâ€ as goal states;in no specific order, 
write a customized uniform cost search algorithm to generate a path that let a visitor visit all those
goal states preserving the local optimum

```python
import heapq

def customized_uniform_cost_search(graph, start, goals):
    """
    Find a path that visits all the goal states while preserving the local optimum.
    
    :param graph: a dictionary representing the graph
    :param start: the starting node
    :param goals: a set of goal nodes
    :return: a list of nodes representing the path, or None if no path is found
    """
    # initialize the frontier with the starting node
    frontier = [(0, [start])]
    # initialize the explored set as an empty set
    explored = set()
    # iterate while the frontier is not empty
    while frontier:
        # remove the node with the lowest cost from the frontier
        (cost, path) = heapq.heappop(frontier)
        # get the last node in the path
        node = path[-1]
        # check if it is a goal node
        if node in goals:
            # if all goal nodes have been visited, return the path
            if set(path) >= goals:
                return path
        # add the node to the explored set
        explored.add(node)
        # expand the node's neighbors
        for neighbor, neighbor_cost in graph.get(node, {}).items():
            # check if the neighbor has already been explored
            if neighbor not in explored:
                # calculate the total cost of the path to the neighbor
                total_cost = cost + neighbor_cost
                # create a new path to the neighbor
                new_path = path + [neighbor]
                # add the new path to the frontier
                heapq.heappush(frontier, (total_cost, new_path))
    # if no path is found, return None
    return None

path = customized_uniform_cost_search(graph_ETUCS, "AddisAbaba", {'Assella', 'Assasa', 'Dodolla', 'Bale', 'SofOumer' })
print(path)
# ['AddisAbaba', 'Adama', 'Assella', 'Assasa', 'Dodolla', 'Bale', 'Goba', 'SofOumer']
```

Assume an adversary joins the Traveling Ethiopia Search Problem as shown in Figure 4. The goal
of the agent would be to reach to a state where it gains good quality of Coffee. Write a class that
shows how MiniMax search algorithm directs an agent to the best achievable destination.

```python
terminal = ["Shambu", "Fincha", "Gimbi", "Limu", "Hosana", "Durame", "Benchi Naji", "Tepi", "Kaffa", "Dilla", "Chiro", "Harar"]
utility = [4, 5, 8, 8, 6, 5, 5, 6, 7, 9, 6, 10]
class MiniMaxAgent:
    def __init__(self, graph, terminal, utility):
        self.graph = graph
        self.terminal = terminal
        self.utility = utility

    def minimax(self, node, depth, maximizingPlayer):
        if node in self.terminal or depth == 0:
            return self.utility[self.terminal.index(node)]
        
        if maximizingPlayer:
            maxEval = float('-inf')
            for child in self.graph[node]:
                eval = self.minimax(child, depth-1, False)
                maxEval = max(maxEval, eval)
            return maxEval
        else:
            minEval = float('inf')
            for child in self.graph[node]:
                eval = self.minimax(child, depth-1, True)
                minEval = min(minEval, eval)
            return minEval

    def get_best_move(self, current_node, depth):
        best_move = None
        best_eval = float('-inf')
        for child in self.graph[current_node]:
            eval = self.minimax(child, depth-1, False)
            if eval > best_eval:
                best_eval = eval
                best_move = child
        return best_move
    
graph = {
    "Addis Ababa": {"Ambo", "Buta Jirra", "Adama"},
    "Ambo": {"Gedo", "Nekemte"},
    "Buta Jirra": {"Worabe", "Wolkite"},
    "Adama": {"Dire Dawa", "Mojo"},
    "Gedo": {"Shambu", "Fincha"},
    "Nekemte": {"Gimbi", "Limu"},
    "Worabe": {"Hosana", "Durame"},
    "Wolkite": {"Benchi Naji", "Tepi"},
    "Mojo": {"Dilla", "Kaffa"},
}

terminal = ["Shambu", "Fincha", "Gimbi", "Limu", "Hosana", "Durame", "Benchi Naji", "Tepi", "Kaffa", "Dilla", "Chiro", "Harar"]
utility = [4, 5, 8, 8, 6, 5, 5, 6, 7, 9, 6, 10]

agent = MinimaxAgent(graph, terminal, utility)
best_move = agent.get_best_move("Addis Ababa", 3)
print("Best move:", best_move)
```
